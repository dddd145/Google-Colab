# **テキストハクスラ**




## **概要**

このプログラムは、テキストのみでハクスラを遊べるようにしたプログラムです。敵は倒すごとに強くなり、落とすお金も増えます。戦闘終了後、ショップでお金を消費し、回復薬や装備ガチャを回せます。無限に続きます。

## **操作方法**

`input()`で入力を求めるので、文章の指示に従い操作します。

## **制作者のコメント**

現在学習済みの知識を生かしてゲームを作ってみました。簡単なコードでのみ制作しました。コード自体は難しくなく、武器の強さと敵の強さのバランスが難しかったです。まだまだ基礎的な知識しか持ち合わせていないので、これからの学習を通して良い作品を制作できるように精進します。

## **作成時間**

作成時間：**約30時間**

```python
%reset -f
import random as r
import time as t
import IPython.display as dis
#変数
money = 10000000
basic_myhp = 10
basic_enemyhp = 5
enemy_lv = 1
cal_reward = 10*(int(enemy_lv/5)*1.5)
diff = 1
power = 1
heal = 0
sword = ["ナマクラ","スチール","ブラック","ライト","デーモン","グレート","エンバー","インフェルノ","レイジ","エレメンタル","ライフ","ゴッド","ここまできたらゲームクリア"]
sword_lv = 0
#入力
def input_command():
  return input()

#SHOP内容
def shopping():
  global money,sword_lv,power,heal
  dis.clear_output()
  print(f"SHOPに来た。何を買おう？現在のお金:\{money}")
  print(f"＿＿＿＿＿＿＿＿＿＿＿＿＿＿＿＿＿＿＿\n\nENTER : メニューに戻る\n1を押してENTER : 剣のグレードを上げる({sword[sword_lv]}=>{sword[sword_lv+1]})[攻撃力: {power} => {power+3}] : \{sword_lv+1*150}\n2を押してENTER : 回復薬を買う(現在の回復薬：{heal}個) : \{basic_enemyhp*10}\n＿＿＿＿＿＿＿＿＿＿＿＿＿＿＿＿＿＿＿")
  if input_command() == 1:
    if money >= sword_lv+1*150:
      power += sword_lv * 3
      money -= sword_lv+1*150
      sword_lv += 1
      print(f"{sword[sword_lv]}ソードになった！")
      input_command()
      shopping()
    else:
      print("お金が足りない")
      input_command()
      shopping()
  elif input_command() == 2:
    print("回復薬を買った")
    heal += 1
  else:
    menu()

#MENU内容
def menu():
  global money
  dis.clear_output()
  print(f"現在のお金:\{money}\n==============================\nENTER : 敵と戦う\n1+ENTER : SHOPにいく\n==============================")
  if input_command() == "":
    dis.clear_output()
    appear_enemy()
  else:
    shopping()

#バトル内容mein
def battle_command(sen_command,can_attack,can_heal):
  global heal
  if heal > 0:
    can_heal = True
  else:
    can_heal = False
  if can_attack == True:
    sen_command =f"・ENTERで攻撃(現在の攻撃力:{power})"
  if can_heal == True:
    sen_command +="\n・1を押してENTERで回復してから攻撃"
  if sen_command == "":
    print("なにもできない")
    enemy_attack()
  t.sleep(0.3)
  print(f"==============================\n{sen_command}\n==============================")
  enc_command = input_command()
  if enc_command == "":
    my_attack()
  else:
    heal -= 1
    myhp += 10
    my_attack()

#バトル内容次は自分のターン
def myturn():
  global myhp
  print(f"次は自分のターン！(自分の体力:{myhp},敵の体力:{enemyhp})")
  battle_command("",True,True)

#バトル内容敵からの攻撃
def enemy_attack():
  global myhp
  myhp -= int(enemy_lv*1.15)
  print(f"敵からの攻撃！{int(enemy_lv*1.15)}のダメージ\n")
  if myhp > 0:
    t.sleep(0.3)
    myturn()
  else:
    print("体力が0になってしまった")

#バトル内容自分の攻撃
def my_attack():
  global enemyhp,money,enemy_lv
  enemyhp -= power
  if enemyhp > 0:
    print(f"{sword[sword_lv]}ソードで攻撃！次は敵のターン！")
    t.sleep(0.3)
    enemy_attack()
  else:
    print(f"{sword[sword_lv]}ソードで攻撃！敵を倒した！")
    t.sleep(0.8)
    reward = r.randint(cal_reward+10*diff,cal_reward*3+30*diff)
    money += reward
    print(f"\{reward} 手に入れた\nENTERで進む")
    enemy_lv += 1
    input_command()
    menu()

#バトル内容エンカウント
def appear_enemy():
  global myhp,enemyhp,heal
  enemyhp = basic_enemyhp + enemy_lv*3
  myhp = basic_myhp
  sen_command=""
  can_attack = True
  if heal > 0:
    can_heal = True
  else:
    can_heal = False
  print(f"敵(Lv.{enemy_lv})が現れた！(自分のHP:{myhp},敵のHP:{enemyhp})")
  battle_command(sen_command,can_attack,can_heal)
  enc_command = input_command()
  if enc_command == "":
    my_attack()
  else:
    heal -= 1
    myhp += 10
    my_attack()

#最初のプログラム
print("現在の難易度の設定は「普通」です。「簡単」に変更しますか？\n1を入力してENTERで変更します。\n変更しなければ何も入力せずENTERで「普通」のままとなります。\n\033[31m※後に変更できません\033[0m\n難易度「簡単」は敵を倒した時の報酬が増えます。")
if input_command() == 1:
  diff = 2
else:
  diff = 1
dis.clear_output()
menu()
```
